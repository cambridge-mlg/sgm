```python
import jax
jax.config.update('jax_platform_name', 'cpu')
from jax import numpy as jnp
import matplotlib.pyplot as plt

# make matplotlib use latex rendering
plt.rc('text', usetex=True)
plt.rc('font', family='serif')
# make latex use amsmath package
plt.rc('text.latex', preamble=r'\usepackage{amsmath}')
```

```python
angles_0_0 = jnp.load('results_old/angles_0_0.npy')
angles_15_15 = jnp.load('results_old/angles_-15_15.npy')
angles_30_30 = jnp.load('results_old/angles_-30_30.npy')
angles_45_45 = jnp.load('results_old/angles_-45_45.npy')
angles_90_90 = jnp.load('results_old/angles_-90_90.npy')
angles_180_180 = jnp.load('results_old/angles_-180_180.npy')
```

```python
# make a matplotlib figure, 400 dpi, to compare the different histograms
fig = plt.figure(figsize=(4, 3), dpi=400)
# add histogram for the angles from 0 to 0
offset_0_0 = 130
plt.hist(angles_0_0 + offset_0_0, bins=30, alpha=0.75, label='$\\pm0^\circ$', histtype='step', density=False)
# add histogram for the angles from -15 to 15
offset_15_15 = 50
# plt.hist(angles_15_15 + offset_15_15, bins=30, alpha=0.75, label='$\\pm15^\circ$', histtype='step', density=False)
# add histogram for the angles from -30 to 30
offset_30_30 = 60
plt.hist(angles_30_30 + offset_30_30, bins=30, alpha=0.75, label='$\\pm30^\circ$', histtype='step', density=False)
# add histogram for the angles from -45 to 45
offset_45_45 = 70
plt.hist(angles_45_45 + offset_45_45, bins=30, alpha=0.75, label='$\\pm45^\circ$', histtype='step', density=False)
# add histogram for the angles from -90 to 90
offset_90_90 = 100
plt.hist(angles_90_90 + offset_90_90, bins=30, alpha=0.75, label='$\\pm90^\circ$', histtype='step', density=False)
# add histogram for the angles from -180 to 180
# offset_180_180 = 0
# plt.hist(angles_180_180 + offset_180_180, bins=30, alpha=0.75, label='$\\pm180^\circ$', histtype='step', density=False)

plt.legend(title='rotation')
plt.xlim(-100, 100)
plt.xlabel('$\\eta \\sim q_{\\boldsymbol\\nu\\!}\\left(\\eta \\,\\middle|\\, \\mathbf{x} \\right)$')
plt.ylabel('count')
plt.show()
fig.savefig('results_old/angles.pdf', bbox_inches='tight')
```

```python
lower_q = 0.05
upper_q = 1 - lower_q
q_diff_0_0 = jnp.quantile(angles_0_0, upper_q) - jnp.quantile(angles_0_0, lower_q)
print("0", q_diff_0_0/2)
q_diff_15_15 = jnp.quantile(angles_15_15, upper_q) - jnp.quantile(angles_15_15, lower_q)
print("15", q_diff_15_15/2)
q_diff_30_30 = jnp.quantile(angles_30_30, upper_q) - jnp.quantile(angles_30_30, lower_q)
print("30", q_diff_30_30/2)
q_diff_45_45 = jnp.quantile(angles_45_45, upper_q) - jnp.quantile(angles_45_45, lower_q)
print("45", q_diff_45_45/2)
q_diff_90_90 = jnp.quantile(angles_90_90, upper_q) - jnp.quantile(angles_90_90, lower_q)
print("90", q_diff_90_90/2)
q_diff_180_180 = jnp.quantile(angles_180_180, upper_q) - jnp.quantile(angles_180_180, lower_q)
print("180", q_diff_180_180/2)
```
